```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Temel Çizim İşlemleri

x eksenine göre y eksenini çiziyoruz. plot(*args[scalex, scaley, data])

```python
plt.plot([1,2,3,4],[1,4,9,16]);
```

```python
x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y)
plt.show() #noktali virgul yerine bu sekilde de bilgi kısmını kapatabiliriz.
```

```python
plt.title("Ilk grafigimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.plot(x,y);
```

```python
#  Normalde centikler otomatik olarak yerlestirilir. Bunu kendimiz de su sekilde yapabiliriz.

plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title("Ilk grafigimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.plot(x,y);
```

### Lejant Koyma

```python
plt.plot(x,y, label = "x^2" , color= "green")
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title("Ilk grafigimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend();
# plt.plot(x,y); gosterilmesi icin plt.legend() komutunu kullaniyoruz.
```

```python
plt.plot(x,y, label = "x^2" , color= "blue" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title("Ilk grafigimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend();
```

### Aynı grafikte birden fazla cizim yapmak

```python
plt.plot(x,y, label = "x^2" , color= "blue" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title("Ilk grafigimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")

# 2.cizgimizi ciziyoruz.

x2 = np.arange(0,5,0.5)
plt.plot(x2, x2*2, label = "2*x", color = "red", linewidth = 2, linestyle = "--", marker = ".")

plt.legend();
```

### Grafigi Bilgisayara Kaydetmek 

```python
plt.plot(x,y, label = "x^2" , color= "blue" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title("Ilk grafigimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")

# 2.cizgimizi ciziyoruz.

x2 = np.arange(0,5,0.5)
plt.plot(x2, x2*2, label = "2*x", color = "red", linewidth = 2, linestyle = "--", marker = ".")

plt.legend();

plt.savefig("ilkgrafik.jpeg", dpi = 300)

# istedigimiz dosya turuyle kaydedebiliyoruz.
```

## Barchart

```python
x = ["Bursa", "Istanbul", "Ankara"]
y = 120, 150, 98

plt.bar(x,y);
```

```python
#  Eger cubuklardan birini digerlerinden farklilastirmak icin boyamak isterseniz:

cubuklar = plt.bar(x,y)
cubuklar[0].set_hatch("/")
cubuklar[1].set_hatch("--")
cubuklar[2].set_hatch("*")
```
